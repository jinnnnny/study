## 2.시각화

### 1) 산점도 : plt.scatter, sns.scatterplot, sns.jointplot

① iris의 Petal.Length와 Petal.Width의 관계를 살펴보기 위해 산점도를 그려봅시다.
* plt.scatter

```python
plt.scatter(iris['Petal.Length'],iris['Petal.Width'])
plt.show()
```

* sns.scatterplot

```python
sns.scatterplot(data = iris, x = 'Petal.Length', y='Petal.Width')
plt.show()
```

* sns.scatterplot(    , hue = 'Species')

```python
sns.scatterplot(data = iris, x = 'Petal.Length', y='Petal.Width',hue = 'Species')
plt.show()
```

* sns.joinplot

```python
sns.jointplot(data = iris, x = 'Petal.Length', y='Petal.Width',hue = 'Species')
plt.show()
```

* 그래프로 부터 파악된 내용을 적어 봅시다.

## 3.수치화 : 상관분석

```python
import scipy.stats as spst
```

① boston.indus와 boston.medv의 관계를 수치화 해 봅시다.  

결과를 해석해 봅시다.

```python
spst.pearsonr(boston['indus'], boston['medv'])
```

② diamond의 각 변수들 간에 상관 계수를 구해 봅시다.

```python
diamond.corr()
```
